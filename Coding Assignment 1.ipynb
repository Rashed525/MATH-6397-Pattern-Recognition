## Spectrogram code for a chirp signal##

### The Gabor transform is given by: $$G(f)(t,\omega)=\hat{f_g}(t,\omega)=\int_{-\infty}^{\infty} f(\tau) e^{-i\tau\omega} \bar{g}(t-\tau)\,d\tau$$###

```python
import numpy as np   #packages
import matplotlib.pyplot as plt
import sounddevice as sd
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 18})

dt = 0.001 
t = np.arange(0,2,dt)
f0 = 50
f1 = 250
t1 = 2
x = np.cos(2*np.pi*t*(f0 + (f1-f0)*np.power(t,2)/(3*t1**2))) 
#plt.plot(x)
#plt.show()

plt.specgram(x, NFFT=128, Fs=1/dt, noverlap=120,cmap='jet') 
#plt.plot(x)
plt.colorbar()
plt.xlabel("Time, s")
plt.ylabel("Power/frequency(dB/Hz)")
plt.show()
```

## Spectrogram of Beethoven's Sonata Pathetique##

```python
import numpy as np
import matplotlib.pyplot as plt   #packages
import sounddevice as sd
import os
import librosa
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 18})


y, sr = librosa.load(os.listdir()[2])
plt.specgram(y[0:1000000], NFFT=5000, Fs=sr, noverlap=400,cmap='jet') 
plt.colorbar()
plt.xlabel("Time, s")
plt.ylabel("Power/frequency(dB/Hz)")
plt.show()
```
